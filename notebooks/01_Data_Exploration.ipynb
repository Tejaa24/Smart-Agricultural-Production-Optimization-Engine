```python
# Crop Distribution

plt.figure(figsize=(15,6))

sns.countplot(data=df, x="label", palette="viridis")

plt.title("Distribution of Crop Types", fontsize=16)
plt.xlabel("Crop")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(15,6))

sns.countplot(
    data=df,
    x="label",
    hue="label",
    palette="viridis",
    legend=False
)

plt.title("Distribution of Crop Types", fontsize=16)
plt.xlabel("Crop")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(10, 8))

correlation = df.drop("label", axis=1).corr()

sns.heatmap(
    correlation,
    annot=True,
    cmap="YlGnBu",
    linewidths=0.5,
    fmt=".2f"
)

plt.title("Correlation Heatmap of Features", fontsize=16)

plt.show()
```

## Observation

- The heatmap shows the correlation between all numerical features.
- Most features have weak to moderate correlations.
- No pair of features shows extremely high correlation, which suggests there is no severe multicollinearity.
- The dataset is suitable for training machine learning models.

```python
# Histograms for Numerical Features

df.hist(figsize=(15, 10), bins=20)

plt.suptitle("Distribution of Numerical Features", fontsize=16)

plt.tight_layout()

plt.show()
```

## Observation

- The histograms show the distribution of all numerical features.
- Most features have a well-distributed range of values.
- The dataset covers diverse environmental conditions, which is beneficial for machine learning.

```python
# Boxplots for Numerical Features

plt.figure(figsize=(15, 6))

sns.boxplot(data=df.drop("label", axis=1))

plt.title("Boxplots of Numerical Features", fontsize=16)

plt.xticks(rotation=45)

plt.show()
```

## Observation

- Boxplots help identify the spread of values and potential outliers.
- Some features contain natural outliers, which are expected in agricultural datasets.
- No abnormal data distribution is observed.
