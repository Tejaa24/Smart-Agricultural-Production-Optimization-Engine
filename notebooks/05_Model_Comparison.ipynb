```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.DataFrame({

    "Model": [
        "Logistic Regression",
        "Decision Tree",
        "Random Forest",
        "KNN",
        "SVM",
        "Naive Bayes"
    ],

    "Accuracy": [
        0.972727,
        0.979545,
        0.995455,
        0.979545,
        0.984091,
        0.995455
    ]
})


results
```

```python
results = results.sort_values(
    by="Accuracy",
    ascending=False
)


results
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    results["Model"],
    results["Accuracy"]
)

plt.xticks(
    rotation=45
)

plt.xlabel("Machine Learning Model")

plt.ylabel("Accuracy")

plt.title(
    "Crop Recommendation Model Performance Comparison"
)

plt.show()
```

## Model Selection Conclusion

After evaluating multiple machine learning algorithms,
Random Forest and Naive Bayes achieved the highest accuracy.

Random Forest was selected as the final model because it:

- Handles complex agricultural relationships effectively.
- Provides robust predictions.
- Performs well on unseen data.
- Is suitable for real-world deployment.
